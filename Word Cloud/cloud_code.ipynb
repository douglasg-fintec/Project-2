```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('word_cloud_data.csv', index_col=None)
df.head()
```

```python
df_words= df[['Top words','frequency']]
df_hashtags= df[['Top hashtags','frequency.1']]
df_hashtags = df_hashtags.dropna()
display(df_words.tail(5))
display(df_hashtags.tail(5))
```

```python
top_words = df_words.set_index('Top words').to_dict()['frequency']
top_hashtags = df_hashtags.set_index('Top hashtags').to_dict()['frequency.1']
```

```python
wc_words = WordCloud(width=800, height=400, max_words=200, background_color='white').generate_from_frequencies(top_words)
wc_tags = WordCloud(width=800, height=400, max_words=200, background_color='white').generate_from_frequencies(top_hashtags)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc_words, interpolation='bilinear')
plt.axis('off')
plt.show()
plt.savefig('words_wc.png')
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc_tags, interpolation='bilinear')
plt.axis('off')
plt.show()
plt.savefig('hash_wc.png')
```
