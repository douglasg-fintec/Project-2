```python
def strip_tweets(input_txt):

   #remove mentions
   pattern1 = r'@[A-Za-z0-9]+' 

    #remove hashtags
   pattern2 = r'#[A-Za-z0-9]+'

   # remove URLs
   pattern3 = r'https?://[A-Za-z0-9./]+' 

   combined_pattern = r'|'.join((pattern1, pattern2, pattern3))
   input_txt = re.sub(combined_pattern, '',input_txt)

   return input_txt

```

```python
def remove_html(input_txt):
    input_txt = BeautifulSoup(input_txt).get_text()
    return input_txt
```

```python
def remove_punctuation(input_txt):


# initializing punctuations string
  punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 # Removing punctuations in string
# Using loop + punctuation string
  for ele in input_txt:
    if ele in punc:
        out_txt = input_txt.replace(ele, '')
    return out_txt

```

```python

 def remove_pattern(input_txt):
    r = re.findall(r"@(w+)", input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)      
    return input_txt
```

```python

clean_tweets = []

for tweet in tweets_df['Tweet']:

#remove the first character of the string
# Remove first character from string if matches 'b'
 tweet = tweet.lstrip('b')

# Remove Re-Tweet first character from string if matches RT
 tweet = tweet.lstrip("RT :")


#Strips inheriant characters in tweets
 stripped = strip_tweets(tweet)

 #remove html characters
 stripped = remove_html(stripped)

 # Remove  punctuation
 striped = remove_punctuation(stripped)

 # make all letters lower case - avoids case sensitive issues with the model
 stripped = stripped.lower()
 
 #cleaned = stripped.''.decode("utf-8-sig").replace(u"\ufffd", "?")
 clean_tweets.append(stripped)
  
```

```python
clean_tweets_df = pd.DataFrame(clean_tweets,columns=['tweet'])
clean_tweets_df['date'] = tweets_df['Date']
clean_tweets_df.head()
```
