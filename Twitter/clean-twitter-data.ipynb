```python
# Import the required libraries and dependencies
import os
import re
from pathlib import Path
import pandas as pd
from dotenv import load_dotenv
from bs4 import BeautifulSoup
import re
```

```python
twitter_df = pd.read_csv(
    Path('../TWITTER/twitter_data.csv'),
    header = None,
    parse_dates=True,
    infer_datetime_format=True).dropna()

twitter_df.head()
```

```python
tweets_df = twitter_df.iloc[:,[3,0]]
tweets_df.columns = ['Date','Tweet']

tweets_df
```

```python
def remove_mentions(input_txt):
 #remove mentions
   pattern = r'@[A-Za-z0-9]+' 
   input_txt = re.sub(pattern, '',input_txt)     
   return input_txt
```

```python
def remove_hash_tags(input_txt):
 #remove mentions
   pattern = r'#[A-Za-z0-9]+' 
   input_txt = re.sub(pattern, '',input_txt)     
   return input_txt
```

```python
def remove_urls(input_txt):
 #remove mentions
   pattern = r'https?://[A-Za-z0-9./]+' 
   input_txt = re.sub(pattern, '',input_txt)     
   return input_txt
```

```python
def remove_special_characters(input_txt):
 #remove mentions
   pattern = r'[!@#%^&*(),.?":{}|_<>]'
   input_txt = re.sub(pattern, '',input_txt)     
   return input_txt

   #[!@#$%^&*(),.?":{}|_<>]
```

```python
def remove_html(input_txt):
    input_txt = BeautifulSoup(input_txt).get_text()
    return input_txt
```

```python
def remove_retweet(input_txt):
 #remove mentions
   pattern = r'^[RT \s]+' 
   input_txt = re.sub(pattern, '',input_txt)     
   return input_txt
```

```python

clean_tweets = []

for tweet in tweets_df['Tweet']:
  
#remove the first character of the string
# Remove first character from string if matches 'b'
  tweet = tweet.lstrip('b')

 #Remove mentions
  tweet = remove_mentions(tweet)

   #Remove Hash Tags
  tweet = remove_hash_tags(tweet)

  # Remove Hash Tags
  tweet = remove_urls(tweet)

  tweet = remove_special_characters(tweet)

 #remove html characters
  tweet = remove_html(tweet)

 #Remove the retweet symbol
  tweet = remove_retweet(tweet)


 # make all letters lower case - avoids case sensitive issues with the model
  tweet = tweet.lower()
 
  cleaned = tweet.decode("utf-8-sig").replace(u"\ufffd", "?")
  clean_tweets.append(tweet)
  
```

```python
clean_tweets_df = pd.DataFrame(clean_tweets,columns=['tweet'])
clean_tweets_df['date'] = tweets_df['Date']
clean_tweets_df
```

```python
output = Path('../TWITTER/twitter_data_CLEANED.csv')
clean_tweets_df.to_csv(output)
```
